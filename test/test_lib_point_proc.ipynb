```python
%matplotlib inline
import matplotlib.pyplot as plt
from os import sys
import numpy as np
sys.path.append("../")
#sys.path.append("./lib_point_proc")

from basic_models import HomogeneousPoissonProcessModel
np.random.seed(0)

hpp = HomogeneousPoissonProcessModel(l_0=0.1)
T   = hpp.sample(t_max=1000)
hpp.add_realization("all",T)
hpp.set_cur_realization_key("all")
hpp.plot()
hpp.fit_mle_params()
hpp.print_params()
print hpp.coef_variation()
print hpp.iei_corr()["pearsonr"]
ax = plt.figure().add_subplot(111)
hpp.qq_plot(ax=ax)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from os import sys
import numpy as np
sys.path.append("./lib_point_proc")

from univariate_hawkes import UnivariateHawkesProcessModel
from basic_models import HomogeneousPoissonProcessModel

np.random.seed(0)

hawkes = UnivariateHawkesProcessModel(l_0=0.1,alpha=0.1,beta=0.2)
T = hawkes.sample(t_max=500)
hawkes.add_realization("all",T)
hawkes.set_cur_realization_key("all")
hawkes.plot()
print hawkes.log_likelihood()
hawkes.fit_mle_params()
hawkes.print_params()

hawkes.fit_map_params()
hawkes.print_params()


print hawkes.coef_variation()
print hawkes.iei_corr()

hp = HomogeneousPoissonProcessModel()
hp.add_realization("all",T)
hp.set_cur_realization_key("all")
hp.fit_mle_params()
print hp.log_likelihood()
#ax = plt.figure().add_subplot(111)
#hawkes.qq_plot(ax)
#hp.qq_plot(ax)
```

1. Hawkes Process(HP) is generalization of Homogeneous PP (HPP). HP can fit the data generated from HPP correctly but the reverse is not true.

2. Coefficient of Variation (cov) = $\frac{\sigma}{\mu}=1 $ for exponential distribution. Hence for data from HPP cov must be close to 1. and for HP it must be far from 1. 

3. The correlation coefficient should be high with low p value for nhpp e.g. hawkes process
